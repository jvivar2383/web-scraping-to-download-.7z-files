## Web scrapping function to download  .7z files

```python
from bs4 import BeautifulSoup
import datetime
import matplotlib.pyplot as plt
import logging
import numpy as np
import os
import pandas as pd
import py7zr
import requests
import shutil
import ssl
import urllib
import warnings
import xarray as xr
#import pycurl as curl
import re
from requests.packages.urllib3.util.retry import Retry
from requests.adapters import HTTPAdapter
```

```python
def web_file_scrapper(url, tag = 'a'):
    
    
    
    """
    this function takes as input an url with a page of nested windows 7 files (.7z). each zipfile 
    is downloaded and save locally with the name of the href link. The function also takes a tag input reepresenting 
    the tag you want to search for in the html file. The default argument is set to 'a' for anchor because is the 
    most common. The function checks if the file is already in the system to avoid double downloads of the file.
    
    """
    #workaround to prevent certification
    ssl._create_default_https_context = ssl._create_unverified_context 
    html_page = requests.get(url, verify=False).text
    soup = BeautifulSoup(html_page, features='html.parser')

    anchor = soup.find_all(tag)
    count =0
    count1 = 0
  
    for a in anchor:
        
        href= a.get('href')
        
        #searches for the files we want only withing the anchors
        match = re.search("20..........0S",href )
        
        if match:
               
            url2 = url + href
            
            ssl._create_default_https_context = ssl._create_unverified_context
            html_page2 = requests.get(url2, verify = False, stream = False)
            soup_of_day = BeautifulSoup(html_page2.text, features = "html.parser")
        
            anchor2 = soup_of_day.find_all(tag)
            
            for link in anchor2:
                ssl._create_default_https_context = ssl._create_unverified_context 
                href2 = link.get('href')
                print(f"This is the link value: {link}")
                
                # Returns boolean representing whether or not the file exists
                match2 = re.search("Com.............7z", href2)
                
                if match2:
                    #ssl._create_default_https_context = ssl._create_unverified_context
                    url3 = url2+href2
                    
                    filename = os.path.basename(url3)
                    
                    #checks if file exist in the current directory
                    file_exists = os.path.exists(filename)
                    
                    if file_exists != True:
                        print(f"saving file {filename} in directory")
                        ssl._create_default_https_context = ssl._create_unverified_context 
                        response = requests.get(url3, stream = False, verify = False, timeout = 10)

                        ssl._create_default_https_context = ssl._create_unverified_context
                        with open(filename, 'wb') as out:
                            out.write(response.content)
                    else:
                        print(f"file {filename} already in repertory")
                else:
                    count1 +=1
                    print(f"file did not have a .7z file in it. Total ommited files: {count1}")
        else:
            count +=1
            print(f"no files ending in 0S. Files disregarded from main page: {count}")
            
```

```python
url = "https://datadb.noaacrest.org/public/ORSL/Archived_Lidar/200S/2021/"
files = web_file_scrapper(url = url, tag='a')
```
